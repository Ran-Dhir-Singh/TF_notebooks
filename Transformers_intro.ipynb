```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis", model='ProsusAI/finbert')
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
classifier("Its raining today since morning.")
```

```python
classifier("Its raining today since morning and I have to go out to meet somone")
```

```python
classifier("Its a sunny day so iI can play cricket with my friends after a long time.")
```

```python
classifier("i scored well in my exams.")
```
